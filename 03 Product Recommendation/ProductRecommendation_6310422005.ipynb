```python
import pandas as pd

df = pd.read_csv('Customer Preference Survey (Responses) - Form Responses 1.csv')
```

```python
df.head(5)
```

```python
df = df.replace("ไม่เคยซื้อ",0)
df = df.replace("ไม่เคยซือ",0)
df = df.replace("ไม่เคย",0)
df = df.replace("ไม่",0)
df = df.replace("เคยซื้อ",1)
df = df.replace("เคย",1)
df = df.fillna(0)
#df['Transaction'] = df.index+1
df = df.drop(columns="Timestamp")
df
```

```python
#check missing
missing = df.isnull().sum()
print(missing)
```

```python
#hot_encoded_df = df.set_index(['Transaction'])
```

```python
#change to integer
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
hot_encoded_df = df.applymap(encode_units)
```

```python
hot_encoded_df.info()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True, max_len=2)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5) &
      (rules['support'] >= 0.5) ].sort_values(by=['lift']).head(20)
```
